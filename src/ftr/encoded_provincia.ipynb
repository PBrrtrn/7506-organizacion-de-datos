```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)

training_data = pd.read_csv('../../data/TP2/train.csv', usecols=['id','provincia'])
evaluation_data = pd.read_csv('../../data/TP2/test.csv', usecols=['id','provincia'])

training_data.head()
```

```python
training_data['provincia'] = training_data['provincia'].apply(lambda x: str(x))
evaluation_data['provincia'] = evaluation_data['provincia'].apply(lambda x: str(x))

training_entities = training_data[['provincia']].to_numpy()
evaluation_entities = evaluation_data[['provincia']].to_numpy()
all_entities = np.unique( np.concatenate( (training_entities, evaluation_entities), axis=None) )
```

```python
for entity in all_entities:
    training_data['entity_'+entity] = training_data['provincia'].apply(lambda x: 1 if x == entity else 0)
    evaluation_data['entity_'+entity] = evaluation_data['provincia'].apply(lambda x: 1 if x == entity else 0)
    
training_data.drop(columns=['provincia'], inplace = True)
evaluation_data.drop(columns=['provincia'], inplace = True)

training_data.head()
```

```python
training_data.to_csv('../../res/ftr/encoded_provincia_train.csv')
evaluation_data.to_csv('../../res/ftr/encoded_provincia_evaluation.csv')
```
