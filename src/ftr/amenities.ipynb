```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)

cols = ['id', 'gimnasio', 'usosmultiples', 'piscina',
        'escuelascercanas', 'centroscomercialescercanos']

training_raw = pd.read_csv('../../data/TP2/train.csv', usecols = cols)
evaluation_raw = pd.read_csv('../../data/TP2/test.csv', usecols = cols)

training_raw.head()
```

```python
def any_amenity(x):
    return float(1) if x >= 1 else 0

training_raw['property_amenities'] = (training_raw.gimnasio + training_raw.usosmultiples + training_raw.piscina)
training_raw['any_property_amenity'] = (training_raw.property_amenities).apply(any_amenity)
training_raw['all_propety_amenities'] = (training_raw.gimnasio * training_raw.usosmultiples * training_raw.piscina)

training_raw['location_amenities'] = (training_raw.escuelascercanas + training_raw.centroscomercialescercanos)
training_raw['any_location_amenity'] = (training_raw.location_amenities).apply(any_amenity)
training_raw['all_location_amenities'] = (training_raw.escuelascercanas * training_raw.centroscomercialescercanos)

training_raw['number_of_amenities'] = training_raw.gimnasio + training_raw.usosmultiples + training_raw.piscina + training_raw.escuelascercanas + training_raw.centroscomercialescercanos
training_raw['any_amenity'] = training_raw['number_of_amenities'].apply(any_amenity)
training_raw['all_amenities'] = training_raw.gimnasio * training_raw.usosmultiples * training_raw.piscina * training_raw.escuelascercanas * training_raw.centroscomercialescercanos

evaluation_raw['property_amenities'] = (evaluation_raw.gimnasio + evaluation_raw.usosmultiples + evaluation_raw.piscina)
evaluation_raw['any_property_amenity'] = (evaluation_raw.property_amenities).apply(any_amenity)
evaluation_raw['all_propety_amenities'] = (evaluation_raw.gimnasio * evaluation_raw.usosmultiples * evaluation_raw.piscina)

evaluation_raw['location_amenities'] = (evaluation_raw.escuelascercanas + evaluation_raw.centroscomercialescercanos)
evaluation_raw['any_location_amenity'] = (evaluation_raw.location_amenities).apply(any_amenity)
evaluation_raw['all_location_amenities'] = (evaluation_raw.escuelascercanas * evaluation_raw.centroscomercialescercanos)

evaluation_raw['number_of_amenities'] = evaluation_raw.gimnasio + evaluation_raw.usosmultiples + evaluation_raw.piscina + evaluation_raw.escuelascercanas + evaluation_raw.centroscomercialescercanos
evaluation_raw['any_amenity'] = evaluation_raw['number_of_amenities'].apply(any_amenity)
evaluation_raw['all_amenities'] = evaluation_raw.gimnasio * evaluation_raw.usosmultiples * evaluation_raw.piscina * evaluation_raw.escuelascercanas * evaluation_raw.centroscomercialescercanos

training_raw.head(10)
```

```python
dropcols = ['gimnasio', 'usosmultiples', 'piscina', 'escuelascercanas', 'centroscomercialescercanos']

training_data = training_raw.drop(columns = dropcols)
evaluation_data = evaluation_raw.drop(columns = dropcols)

training_data.head()
```

```python
training_data.to_csv('../../res/ftr/amenities_train.csv')
evaluation_data.to_csv('../../res/ftr/amenities_evaluation.csv')
```
