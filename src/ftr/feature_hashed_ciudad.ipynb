```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import FeatureHasher

training_data = pd.read_csv('../../data/TP2/train.csv', usecols=['id','ciudad'])
evaluation_data = pd.read_csv('../../data/TP2/test.csv', usecols=['id','ciudad'])

training_data.head()
```

```python
training_data['ciudad'] = training_data['ciudad'].apply(lambda x: str(x))
evaluation_data['ciudad'] = evaluation_data['ciudad'].apply(lambda x: str(x))

feature_hasher = FeatureHasher(n_features = 40, input_type='string')

training_hashes = feature_hasher.transform(training_data['ciudad'])
evaluation_hashes = feature_hasher.transform(evaluation_data['ciudad'])

hashed_cities_training = pd.DataFrame(training_hashes.toarray() ).add_prefix('ciudad_hash_')
hashed_cities_evaluation = pd.DataFrame(evaluation_hashes.toarray() ).add_prefix('ciudad_hash_')

hashed_cities_training['id'] = training_data['id']
hashed_cities_evaluation['id'] = evaluation_data['id']

hashed_cities_training.head()
```

```python
hashed_cities_training.to_csv('../../res/ftr/feature_hashed_ciudad_train.csv')
hashed_cities_evaluation.to_csv('../../res/ftr/feature_hashed_ciudad_evaluation.csv')
```
