```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)

types = {'id': int, 'titulo': str, 'descripcion': str, 'tipodepropiedad': str, 'direccion': str,
        'ciudad': str, 'provincia': str, 'antiguedad': np.float32, 'habitaciones': np.float32,
        'garages':np.float32, 'idzona': np.float32, 'lat': np.float32, 'lng': np.float32,
        'gimnasio': int, 'usosmultiples': int, 'piscina': int, 'escuelascercanas': int,
        'centroscomercialescercanos': int, 'precio':int}

training_raw = pd.read_csv('../../data/TP2/train.csv', dtype = types)
evaluation_raw = pd.read_csv('../../data/TP2/test.csv', dtype = types)

training_raw.head()
```

```python
training_data = pd.DataFrame()
evaluation_data = pd.DataFrame()

training_data['id'] = training_raw['id']
evaluation_data['id'] = evaluation_raw['id']

training_data['metrostotalesporhabitacion'] = training_raw['metrostotales']/training_raw['habitaciones']
evaluation_data['metrostotalesporhabitacion'] = evaluation_raw['metrostotales']/evaluation_raw['habitaciones']

training_data['metroscubiertosporhabitacion'] = training_raw['metroscubiertos']/training_raw['habitaciones']
evaluation_data['metroscubiertosporhabitacion'] = evaluation_raw['metroscubiertos']/evaluation_raw['habitaciones']

training_data['metroscubiertossobretotales'] = training_raw['metroscubiertos']/training_raw['metrostotales']
evaluation_data['metroscubiertossobretotales'] = evaluation_raw['metroscubiertos']/evaluation_raw['metrostotales']

training_data.head()
```

```python
training_data.to_csv('../../res/ftr/surface_features_train.csv')
evaluation_data.to_csv('../../res/ftr/surface_features_evaluation.csv')
```
