```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 30)

training_data = pd.read_csv('../../data/TP2/train.csv', usecols=['id','ciudad'])
evaluation_data = pd.read_csv('../../data/TP2/test.csv', usecols=['id','ciudad'])

training_data.head()
```

```python
training_data['ciudad'] = training_data['ciudad'].apply(lambda x: str(x))
evaluation_data['ciudad'] = evaluation_data['ciudad'].apply(lambda x: str(x))

training_cities = training_data[['ciudad']].to_numpy()
evaluation_cities = evaluation_data[['ciudad']].to_numpy()
all_cities = np.unique( np.concatenate( (training_cities, evaluation_cities), axis=None) )
```

```python
for city in all_cities:
    training_data['city_'+city] = training_data['ciudad'].apply(lambda x: 1 if x == city else 0)
    evaluation_data['city_'+city] = evaluation_data['ciudad'].apply(lambda x: 1 if x == city else 0)
    
training_data.drop(columns=['ciudad'], inplace = True)
evaluation_data.drop(columns=['ciudad'], inplace = True)

training_data.head()
```

```python
training_data.to_csv('../../res/ftr/encoded_ciudad_train.csv')
evaluation_data.to_csv('../../res/ftr/encoded_ciudad_evaluation.csv')
```
