```python
import pandas as pd
import numpy as np
from statistics import mean

pd.set_option('display.max_columns', 30)

types = {'id': int, 'provincia': str}

training_data = pd.read_csv('../../data/TP2/train.csv', dtype = types, usecols=['id', 'tipodepropiedad'])
evaluation_data = pd.read_csv('../../data/TP2/test.csv', dtype = types, usecols=['id', 'tipodepropiedad'])

training_data.head()
```

```python
evaluation_data['tipodepropiedad'] = evaluation_data['tipodepropiedad'].apply(lambda x: str(x))
training_data['tipodepropiedad'] = training_data['tipodepropiedad'].apply(lambda x: str(x))

all_types = ['Apartamento', 'Casa en condominio', 'Casa', 'Terreno',
             'Terreno comercial', 'Local Comercial', 'Quinta Vacacional',
             'Oficina comercial', 'Edificio', 'Casa uso de suelo',
             'Local en centro comercial', 'Bodega comercial', 'Otros', 'Villa',
             'Duplex', 'Inmuebles productivos urbanos',
             'Departamento Compartido', 'Nave industrial', 'Rancho',
             'Terreno industrial', 'nan', 'Huerta', 'Lote', 'Hospedaje',
             'Garage']

for each_type in all_types:
    training_data['type_'+each_type] = training_data['tipodepropiedad'].apply(lambda x: 1 if x == each_type else 0)
    evaluation_data['type_'+each_type] = evaluation_data['tipodepropiedad'].apply(lambda x: 1 if x == each_type else 0)
    
training_data.drop(columns=['tipodepropiedad'], inplace = True)
evaluation_data.drop(columns=['tipodepropiedad'], inplace = True)

training_data.head()
```

```python
training_data.to_csv('../../res/ftr/encoded_tipodepropiedad_train.csv')
evaluation_data.to_csv('../../res/ftr/encoded_tipodepropiedad_evaluation.csv')
```
