```python
import pandas as pd
import numpy as np

types = {'id': int, 'titulo': str, 'descripcion': str, 'tipodepropiedad': str, 'direccion': str,
        'ciudad': str, 'provincia': str, 'antiguedad': np.float32, 'habitaciones': np.float32,
        'garages':np.float32, 'idzona': np.float32, 'lat': np.float32, 'lng': np.float32,
        'gimnasio': int, 'usosmultiples': int, 'piscina': int, 'escuelascercanas': int,
        'centroscomercialescercanos': int, 'precio':int}

non_numerical_types = ['titulo', 'descripcion', 'tipodepropiedad', 'direccion', 'ciudad', 'provincia', 'lat', 'lng','idzona']

pd.set_option('display.max_columns', 30)

training_data = pd.read_csv('../../data/TP2/train.csv', dtype = types, \
                            usecols=['id', 'metroscubiertos', 'metrostotales', 'banos', 'tipodepropiedad','ciudad', 'habitaciones'])
test_data = pd.read_csv('../../data/TP2/test.csv', dtype = types, \
                            usecols=['id', 'metroscubiertos', 'metrostotales', 'banos', 'tipodepropiedad','ciudad', 'habitaciones'])
training_data.head()
```

```python
por_ciudadt = training_data.groupby('ciudad').agg(metros_avg_cubiertos_ciudad = ('metroscubiertos','mean'),\
                                                 metros_avg_ciudad = ('metrostotales','mean'), habitaciones_avg_ciudad = ('habitaciones','mean')\
                                                 ,banios_avg_ciudad = ('banos','mean')).reset_index()

por_ciudad_ev = test_data.groupby('ciudad').agg(metros_avg_cubiertos_ciudad = ('metroscubiertos','mean'),\
                                                 metros_avg_ciudad = ('metrostotales','mean'), habitaciones_avg_ciudad = ('habitaciones','mean')\
                                               ,banios_avg_ciudad = ('banos','mean')).reset_index()
por_ciudadt.head(20)
```

```python
por_ciudad_y_tipot = training_data.groupby(['ciudad','tipodepropiedad']).agg(metros_avg_cub_ciudad_tipo = ('metroscubiertos','mean'),\
                                                 metros_avg_ciudad_tipo = ('metrostotales','mean'), habitaciones_avg_ciudad_tipo = ('habitaciones','mean')\
                                                ,banios_avg_ciudad_tipo = ('banos','mean')).reset_index()

por_ciudad_y_tipo_ev = test_data.groupby(['ciudad','tipodepropiedad']).agg(metros_avg_cub_ciudad_tipo = ('metroscubiertos','mean'),\
                                                 metros_avg_ciudad_tipo = ('metrostotales','mean'), habitaciones_avg_ciudad_tipo = ('habitaciones','mean')\
                                                  ,banios_avg_ciudad_tipo = ('banos','mean')).reset_index()

por_ciudad_y_tipot.head(10)
```

```python
ciudad_tipo_train = training_data.merge(por_ciudad_y_tipot, on = ['ciudad','tipodepropiedad'],how = 'outer')
ciudad_tipo_test = test_data.merge(por_ciudad_y_tipo_ev, on = ['ciudad','tipodepropiedad'],how = 'outer')
ciudad_tipo_train.head()
```

```python
ciudad_train = training_data.merge(por_ciudadt, on = ['ciudad'], how = 'outer')
ciudad_test = test_data.merge(por_ciudad_ev, on = ['ciudad'], how = 'outer')
ciudad_test.head()
```

```python
ciudad_train.drop(columns=['metrostotales','metroscubiertos','habitaciones','ciudad','tipodepropiedad','banos'],inplace=True)
ciudad_test.drop(columns=['metrostotales','metroscubiertos','habitaciones','ciudad','tipodepropiedad','banos'],inplace=True)
ciudad_tipo_train.drop(columns=['metrostotales','metroscubiertos','habitaciones','ciudad','tipodepropiedad','banos'],inplace=True)
ciudad_tipo_test.drop(columns=['metrostotales','metroscubiertos','habitaciones','ciudad','tipodepropiedad','banos'],inplace=True)
ciudad_train.head()
```

```python
ciudad_tipo_train.head()
```

```python
ciudad_tipo_train = ciudad_tipo_train.merge(ciudad_train, on = 'id', how = 'outer')
ciudad_tipo_test = ciudad_tipo_test.merge(ciudad_test, on = 'id', how = 'outer')
ciudad_tipo_train.info()
```

```python
ciudad_tipo_train.to_csv('../../res/ftr/avgs_por_ciudad_y_tipoprop_train.csv')
ciudad_tipo_test.to_csv('../../res/ftr/avgs_por_ciudad_y_tipoprop_evaluation.csv')
```
