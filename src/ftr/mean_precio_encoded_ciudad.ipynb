```python
import pandas as pd
import numpy as np

cols_train = ['id', 'ciudad', 'precio']
cols_evaluation = ['id', 'ciudad']

types = {'id': int, 'ciudad': str, 'precio': float}

training_data = pd.read_csv('../../data/TP2/train.csv', usecols = cols_train, dtype = types)
evaluation_data = pd.read_csv('../../data/TP2/test.csv', usecols = cols_evaluation, dtype = types)

training_data.head()
```

```python
training_data['ciudad'] = training_data['ciudad'].apply(lambda x: str(x) )
evaluation_data['ciudad'] = evaluation_data['ciudad'].apply(lambda x: str(x) )

mean_precios = training_data.groupby('ciudad').agg({'precio':'mean'}).reset_index()

precio_mean_by_ciudad = {}
for tipo in mean_precios['ciudad'].unique():
    precio_mean_by_ciudad[tipo] = mean_precios.loc[mean_precios['ciudad']==tipo]['precio'].values[0]

precio_mean_by_ciudad
```

```python
global_mean = np.mean(list(precio_mean_by_ciudad.values() ) )

training_data['mean_precio_encoded_ciudad'] = training_data['ciudad'].apply(lambda x: precio_mean_by_ciudad.get(x, global_mean) )
evaluation_data['mean_precio_encoded_ciudad'] = evaluation_data['ciudad'].apply(lambda x: precio_mean_by_ciudad.get(x, global_mean))

training_data.drop(columns=['ciudad', 'precio'], inplace = True)
evaluation_data.drop(columns=['ciudad'], inplace = True)

evaluation_data.head()
```

```python
training_data.to_csv('../../res/ftr/mean_precio_encoded_ciudad_train.csv')
evaluation_data.to_csv('../../res/ftr/mean_precio_encoded_ciudad_evaluation.csv')
```
