```python
import pandas as pd
import numpy as np

types = {'id': int, 'titulo': str, 'descripcion': str, 'tipodepropiedad': str, 'direccion': str,
        'ciudad': str, 'provincia': str, 'antiguedad': np.float32, 'habitaciones': np.float32,
        'garages':np.float32, 'idzona': np.float32, 'lat': np.float32, 'lng': np.float32, 'banos':np.float32,
        'metroscubiertos': np.float32, 'metrostotales': np.float32}

dtypes = {'titulo': str, 'descripcion': str, 'tipodepropiedad': str, 'direccion': str,
        'ciudad': str, 'provincia': str, 'antiguedad': np.float32, 'habitaciones': np.float32,
        'garages':np.float32, 'idzona': np.float32, 'lat': np.float32, 'lng': np.float32, 'banos':np.float32,
        'metroscubiertos': np.float32, 'metrostotales': np.float32}

pd.set_option('display.max_columns', 30)

training_data = pd.read_csv('../../data/TP2/train.csv', dtype = types, usecols = types)
test_data = pd.read_csv('../../data/TP2/test.csv', dtype = types, usecols = types)

training_data.head()
```

```python
training_data['banos2'] = training_data['banos'].apply(lambda x: x*x)
training_data['metroscub2'] = training_data['metroscubiertos'].apply(lambda x: x*x)
training_data['habitaciones2'] = training_data['habitaciones'].apply(lambda x: x*x)
training_data['metrostot2'] = training_data['metrostotales'].apply(lambda x: x*x)
training_data['antiguedad2'] = training_data['antiguedad'].apply(lambda x: -x*x)

test_data['banos2'] = test_data['banos'].apply(lambda x: x*x)
test_data['metroscub2'] = test_data['metroscubiertos'].apply(lambda x: x*x)
test_data['habitaciones2'] = test_data['habitaciones'].apply(lambda x: x*x)
test_data['metrostot2'] = test_data['metrostotales'].apply(lambda x: x*x)
test_data['antiguedad2'] = test_data['antiguedad'].apply(lambda x: -x*x)
```

```python
training_data = training_data.drop(columns=dtypes)
test_data = test_data.drop(columns=dtypes)
training_data.head()
```

```python
training_data.to_csv('../../res/ftr/squared_train.csv')
test_data.to_csv('../../res/ftr/squared_data.csv')
```
