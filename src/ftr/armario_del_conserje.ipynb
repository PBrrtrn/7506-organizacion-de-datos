```python
import pandas as pd
import numpy as np

types = {'id': int, 'descripcion': str,'antiguedad': np.float32, 'habitaciones': np.float32,
        'metroscubiertos':np.float32, 'banos': np.float32}

dtypes = {'descripcion': str,'antiguedad': np.float32, 'habitaciones': np.float32,
        'metroscubiertos':np.float32, 'banos': np.float32}

non_numerical_types = ['titulo', 'descripcion', 'tipodepropiedad', 'direccion', 'ciudad', 'provincia', 'lat', 'lng']

pd.set_option('display.max_columns', 30)

training_data = pd.read_csv('../../data/TP2/train.csv', dtype = types, usecols = types)
test_data = pd.read_csv('../../data/TP2/test.csv', dtype = types, usecols = types)
```

```python
training_data.head()
```

```python
training_data['banos_log'] = training_data['banos'].apply(lambda x: np.log(x+1))
training_data['mtrs_banos'] = training_data['metroscubiertos']/training_data['banos']
training_data['mtrs_banos_log'] = training_data['mtrs_banos'].apply(lambda x: np.log(x))
training_data['banos_por_hab'] = training_data['habitaciones']/training_data['banos']
training_data['banos_menos_habitaciones'] = training_data['banos'] - training_data['habitaciones']

test_data['banos_log'] = test_data['banos'].apply(lambda x: np.log(x))
test_data['mtrs_banos'] = test_data['metroscubiertos']/training_data['banos']
test_data['mtrs_banos_log'] = test_data['mtrs_banos'].apply(lambda x: np.log(x))
test_data['banos_por_hab'] = test_data['habitaciones']/training_data['banos']
test_data['banos_menos_habitaciones'] = test_data['banos'] - training_data['habitaciones']
```

```python
training_data = training_data.drop(columns=dtypes)
test_data = test_data.drop(columns=dtypes)
training_data.head()
```

```python
training_data.to_csv('../../res/ftr/armario_del_conserje_train.csv')
test_data.to_csv('../../res/ftr/armario_del_conserje_evaluation.csv')
```
