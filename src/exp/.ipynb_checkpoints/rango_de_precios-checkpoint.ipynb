```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

types = {'id': int, 'titulo': str, 'descripcion': str, 'tipodepropiedad': str, 'direccion': str,
        'ciudad': str, 'provincia': str, 'antiguedad': np.float32, 'habitaciones': np.float32,
        'garages':np.float32, 'idzona': np.float32, 'lat': np.float32, 'lng': np.float32,
        'gimnasio': bool, 'usosmultiples': bool, 'piscina': bool, 'escuelascercanas':bool,
        'centroscomercialescercanos':bool, 'precio':int}
data = pd.read_csv('../../data/train.csv', dtype = types)

data.head()
```

```python
propiedades_por_tipo = pd.DataFrame()
propiedades_por_tipo['tipodepropiedad'] = data['tipodepropiedad']
propiedades_por_tipo['precio_en_miles'] = data['precio']/1000
propiedades_por_tipo.head()
```

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(x = propiedades_por_tipo.tipodepropiedad, y = propiedades_por_tipo.precio_en_miles, width = 1,
           fliersize = 1)
plt.xticks(rotation=75)
```

```python
propiedades_por_provincia = pd.DataFrame()
propiedades_por_provincia['provincia'] = data['provincia']
propiedades_por_provincia['precio_en_miles'] = data['precio']/1000
propiedades_por_provincia.head()
```

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(x = propiedades_por_provincia.provincia, y = propiedades_por_provincia.precio_en_miles, width = 1,
           fliersize = 1)
plt.xticks(rotation=75)
```

```python
propiedades_por_antiguedad = pd.DataFrame()
propiedades_por_antiguedad['antiguedad'] = data['antiguedad']
propiedades_por_antiguedad['precio_en_miles'] = data['precio']/1000
propiedades_por_antiguedad.head()
```

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(x = propiedades_por_antiguedad.antiguedad, y = propiedades_por_antiguedad.precio_en_miles, width = 1,
           fliersize = 1)
plt.xticks(rotation=75)
```

```python
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x = propiedades_por_antiguedad.antiguedad, y = propiedades_por_antiguedad.precio_en_miles)
plt.xticks(rotation=75)
```

```python
propiedades_por_habitaciones = pd.DataFrame()
propiedades_por_habitaciones['habitaciones'] = data['habitaciones']
propiedades_por_habitaciones['precio_en_miles'] = data['precio']/1000
propiedades_por_habitaciones.head()
```

```python
plt.figure(figsize=(8, 4))
ax = sns.boxplot(x = propiedades_por_habitaciones.habitaciones, y = propiedades_por_habitaciones.precio_en_miles, width = 1,
           fliersize = 1, palette = "Blues")
plt.xticks(rotation=75)
```

```python
propiedades_por_anio = pd.DataFrame()
propiedades_por_anio['anio_de_publicacion'] = pd.to_datetime(data['fecha']).apply(lambda x: x.year)
propiedades_por_anio['precio_en_miles'] = data['precio']/1000
propiedades_por_anio.head()
```

```python
plt.figure(figsize=(4, 7))
ax = sns.boxplot(x = propiedades_por_anio.anio_de_publicacion, y = propiedades_por_anio.precio_en_miles, width = 1,
           fliersize = 1, palette = "Blues")
plt.xticks(rotation=75)
```

```python
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x = propiedades_por_anio.anio_de_publicacion, y = propiedades_por_antiguedad.precio_en_miles)
plt.xticks(rotation=75)
```

```python
propiedades_por_mes = pd.DataFrame()
propiedades_por_mes['mes_de_publicacion'] = pd.to_datetime(data['fecha']).apply(lambda x: x.month)
propiedades_por_mes['precio_en_miles'] = data['precio']/1000
propiedades_por_mes.head()
```

```python
plt.figure(figsize=(8, 7))
ax = sns.boxplot(x = propiedades_por_mes.mes_de_publicacion, y = propiedades_por_mes.precio_en_miles, width = 1,
           fliersize = 1)
plt.xticks(rotation=75)
```
